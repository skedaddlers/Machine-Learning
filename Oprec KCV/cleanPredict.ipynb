```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv('predict_case.csv' ,sep='|')
```

```python
df = df.drop(['job_description', 'job_benefits', 'job_title'], axis=1)
```

```python
df.isna().sum()
```

```python
df['experience_level'].fillna(df['experience_level'].mode()[0], inplace=True)
df['company_process_time'].fillna(df['company_process_time'].mode()[0], inplace=True)
df['company_size'].fillna(df['company_size'].mode()[0], inplace=True)
df['company_industry'].fillna(df['company_industry'].mode()[0], inplace=True)
df.isna().sum()
```

```python
label_encoder = LabelEncoder()
columns_to_encode = ['location', 'salary_currency', 'career_level', 'experience_level', 
                     'education_level', 'employment_type', 'job_function', 
                     'company_process_time', 'company_size', 'company_industry']
for column in columns_to_encode:
    df[column] = label_encoder.fit_transform(df[column])
```

```python
for column in columns_to_encode:
    print(f"Mapping untuk kolom {column}:")
    print(df[[column]].drop_duplicates().reset_index(drop=True))
    print("\n")
```

```python
df.to_csv('cleaned_data.csv', index=False)
```
